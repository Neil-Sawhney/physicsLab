```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python

# Data for plotting
d = 1000.0
lamb = 649.0

theta = np.linspace(-2, 2, 100)
delta = (2.0 * np.pi / lamb) * d * np.sin(theta)
bigI = (np.cos(delta/2.0))**2
dovlamsintheta = (d / lamb) * np.sin(theta)
fig, ax = plt.subplots()
fig.patch.set_facecolor("white")
ax.plot(dovlamsintheta, bigI, "r-")
ax.set(xlabel='position(d/lambda)*sin(theta)', ylabel='Intensity(I)',
       title='Youngs Double Slit Intensity')
ax.grid()
fig.text(0, 0, "infitesimally thin slits")
fig.text(0, -.05, "d = 1000 nm; $\lambda$ = 649 nm")
fig.savefig("plot1.png", bbox_inches="tight", dpi=300)
plt.show()
```

```python
# Data for plotting

theta = np.arange(-np.pi/2.0, np.pi/2.0, .0001)
delta = (2.0 * np.pi / lamb) * d * np.sin(theta)
dovlamsintheta = (d / lamb) * np.sin(theta)
fig, ax = plt.subplots()
fig.patch.set_facecolor("white")

bigI = (np.sin(4*delta/2.0)/np.sin(delta/2))**2/(4**2)
ax.plot(dovlamsintheta, bigI, "r-", label="N = 4")
bigI = (np.sin(16*delta/2.0)/np.sin(delta/2))**2/(16**2)
ax.plot(dovlamsintheta, bigI, "b", label="N = 16")
bigI = (np.sin(256*delta/2.0)/np.sin(delta/2))**2/(256**2)
ax.plot(dovlamsintheta, bigI, "k-", label="N = 256")
ax.set(xlabel='position(d/lambda)*sin(theta)', ylabel='$Intensity(I) / I_0$',
       title='Youngs N Slit Intensity')
ax.grid()
ax.legend(loc = "upper right")
fig.text(0, 0, "infetesimally thin slits")
fig.text(0, -0.05, "d = 1000 nm; $\lambda$ = 649 nm")
fig.savefig("plot2.png", bbox_inches="tight", dpi=300)
plt.show()
```

```python

# Data for plotting

theta = np.arange(-np.pi/2.0, np.pi/2.0, .0001)
d = 1500
delta = (2.0 * np.pi / lamb) * d * np.sin(theta)
dovlamsintheta = (d / lamb) * np.sin(theta)
fig, ax = plt.subplots()
fig.patch.set_facecolor("white")

bigI = np.sin((np.pi*d*np.sin(theta))/350)**2/(np.pi *d*np.sin(theta)/350)**2
ax.plot(dovlamsintheta, bigI, "b-", label="$\lambda$ = 350")
bigI = np.sin((np.pi*d*np.sin(theta))/700)**2/(np.pi *d*np.sin(theta)/700)**2
ax.plot(dovlamsintheta, bigI, "r-", label="$\lambda$ = 700")

ax.set(xlabel='position(d/lambda)*sin(theta)', ylabel='$Intensity(I) / I_0$',
       title='Youngs Single Slit Intensity')
ax.grid()
ax.legend(loc = "upper right")
fig.text(0, 0, "infitesimely thin slits")
fig.text(0, -.05, "d = 1500 nm")
fig.savefig("plot3.png", bbox_inches="tight", dpi=300)
plt.show()
```

```python

# Data for plotting

theta = np.arange(-48, 48, .0001)
d =2100 
fig, ax = plt.subplots()
fig.patch.set_facecolor("white")

bigI1 = (np.cos((2.0 * np.pi / 350) * d * np.sin(np.radians(theta))/2.0))**2
ax.plot(theta, bigI1, color = '#BB7DB4', label = "Interference")
d = 700
bigI2 = np.sin((np.pi*d*np.sin(np.radians(theta)))/350)**2/(np.pi *d*np.sin(np.radians(theta))/350)**2
ax.plot(theta, bigI2, color = '#6480A5', label="Diffraction")
bigI =  bigI1*bigI2
ax.plot(theta, bigI, color="#CA4141", linewidth = 3, label="Together")

plt.xticks([-45, -30, -15, 0 ,15,30,45],["-45$\degree$", "-30$\degree$", "-15$\degree$","0", "15$\degree$", "30$\degree$", "45$\degree$"] )

ax.set(xlabel=r"$\theta$", ylabel='$Intensity(I) / I_0$',
       title='Youngs Single Slit Intensity')
ax.grid()
ax.legend(loc = "upper right")
fig.text(0, -0.05, "$\lambda = 350 nm$     slit width = 700 nm    slit seperation = 2100 nm")
fig.savefig("plot4.png", bbox_inches="tight", dpi=300)
plt.show()
```
